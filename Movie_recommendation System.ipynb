```python
# from cast we select only top 3 casts i.e main acst of film
# helper function to get list of top3 casts from cast
def get_cast(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(get_cast)
```

```python
# from crew we want to have a director name
def get_director(obj):
    l=[]
    
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    L=[]
    for y in text.split():
        L.append(ps.stem(y))
    
    return " ".join(L)     # converting list into a string " ".join(x)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
# now ist comparing each vector with other 4806 vectors in angles
# distance is inversaly perportional to similarity
# similarity[0] its showing the similarity for the first vector with 4806 vectors
# similarity near to 1 means those 2 vector are more similar i.e movies are simmilar
```

```python
def recommend(movie):
    movie_idx =new_df[new_df['title']== movie].index[0]
    distances =similarity[movie_idx]
    movies_list =sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

    
```

```python
recommend('Batman Begins')
```
